### using regex i.e regular expressions
#### removes everything except letters only, spaces 
| Regex Symbol | Meaning                              | Example                                            |
| ------------ | ------------------------------------ | -------------------------------------------------- |
| `.`          | Any character                        | `a.c` matches `abc`, `a1c`                         |
| `\d`         | Any digit 0-9                        | `\d\d` matches `12`                                |
| `\w`         | Any letter/number/underscore         | `\w+` matches `hello123`                           |
| `\s`         | Any whitespace (space, tab, newline) | `\s` matches `" "`                                 |
| `^`          | Start of string                      | `^Hello` matches "Hello world"                     |
| `$`          | End of string                        | `world$` matches "Hello world"                     |
| `[^...]`     | Not these characters                 | `[^a-z]` matches anything except lowercase letters |
